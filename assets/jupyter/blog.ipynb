```python
rgb_array = np.array([mpl.colors.to_rgb(color) for color in sns.color_palette('pastel')])
print(rgb_array)
hsv_array = np.array([mpl.colors.rgb_to_hsv(rgb_value) for rgb_value in rgb_array])
print(hsv_array) #hue, saturation, lightness
```

```python
sns.color_palette('pastel')
```

```python
for color in sns.color_palette('pastel'):
    rgb_value = mpl.colors.to_rgb(color)
    hsv_value = np.array(mpl.colors.rgb_to_hsv(rgb_value))
    sns.scatterplot(x=[hsv_value[0]], y=[hsv_value[1]],color=color,legend=None)
```

```python
palette = sns.color_palette('colorblind')
fig = plt.figure()
ax = fig.add_subplot(projection='polar')
for color in palette:
    rgb_value = mpl.colors.to_rgb(color)
    hsv_value = np.array(mpl.colors.rgb_to_hsv(rgb_value))
    ax.scatter(x=[2 * np.pi * hsv_value[0]], 
               y=[hsv_value[1]],color=color)

ax.set_yticks([])
```
