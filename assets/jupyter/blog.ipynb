```python
for color in sns.color_palette('pastel'):
    rgb_value = mpl.colors.to_rgb(color)
    hsv_value = mpl.colors.rgb_to_hsv(rgb_value)
    sns.scatterplot(x=[hsv_value[0]], y=[hsv_value[1]],color=color,legend=None)
```

```python
palette = sns.color_palette('colorblind')
rgb_array, hsv_array = [], []
for color in palette:
    rgb_value = mpl.colors.to_rgb(color)
    hsv_value = np.array(mpl.colors.rgb_to_hsv(rgb_value))
    rgb_array.append(rgb_value)
    hsv_array.append(hsv_value)

rgb_array = np.array(rgb_array)
hsv_array = np.array(hsv_array)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='polar')
ax.scatter(x=2 * np.pi * hsv_array[:, 0], 
           y=hsv_array[:,1],color=[color for color in palette])
ax.set_yticks([])



```
